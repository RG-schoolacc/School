```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/drive/MyDrive/myDATA/DSC502ProjectData/v20.data.curves_post_qc.txt', sep='\t')
#cleans data
df = df[df['apparent_ec50_umol'] > 0]
plt.figure(figsize=(10, 6))
sns.histplot(df['area_under_curve'], bins=50, color='skyblue', kde=True)
plt.title('Distribution of Drug Sensitivity (AUC)')
plt.xlabel('AUC (Lower = More Sensitive)')
#of trials
plt.ylabel('Trials')
plt.show()
#top 5 df
top_5_ids = df.groupby('master_cpd_id')['area_under_curve'].mean().sort_values().head(5).index
df_top5 = df[df['master_cpd_id'].isin(top_5_ids)]
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df_top5, x="area_under_curve", hue="master_cpd_id",element="step", palette="bright", kde=True)
plt.title('Sensitivity Distributions')
plt.xlabel('AUC (Lower = More Sensitive)')
#of trials
plt.ylabel('Trials')
plt.show()
```
